## Read ASS File

```python
with open('captions/Friends.S01E03.chs&eng.ass', 'r', encoding='utf-8-sig') as f:
    doc = ass.parse(f)
```

```python
for x in doc.events:
    utterance = x.text.strip().split('{')
    utt_zh = utterance[1].strip().split('}')[-1]
    utt_en = utterance[-2].strip().split('}')[-1]
    print(utt_zh)
    print(utt_en)
    print('=='*50)
```
