```python
from matplotlib import pyplot as plt
from tqdm import tqdm
```

```python
with open('name_record.txt', 'r', encoding='utf-8') as f:
    data = []
    for x in f:
        x = x.strip().split('\t')
        data.append([
            x[0],
            int(x[1])
        ])
```

```python
data_dict = {}
for x in data:
    data_dict[x[1]] = x
```

```python
print(data_dict)
```

```python
gaps = []
for i in range(len(data)-1):
    # print(data[i][0], data[i][1], data[i+1][1]-data[i][1])
    gaps.append(data[i+1][1]-data[i][1])
```

```python
count = 0
indexs = []
for i in range(len(data)-5):
    signal = max(gaps[i:i+5])
    if signal < 500:
        count += 1
        # print(data[i], signal)
        indexs.append(data[i][1])
        # print('=='*50)
```

```python
print(indexs)
```

```python
final_gaps = []
mini = indexs[0]
maxi = indexs[0]
for i in range(len(indexs)-1):
    item = indexs[i+1]
    if item - maxi < 6000:
        maxi = item
    else:
        if maxi - mini > 10:
            final_gaps.append([mini, maxi])
        mini = item
        maxi = item
```

```python
print(len(final_gaps))
```

```python
all = 0
for item in final_gaps:
    all += (item[1] - item[0])
    print(item, item[1]-item[0])
    print(data_dict[item[0]])
    print(data_dict[item[1]])
    print('=='*50)
```

```python
print(all)
```

```python
areas = []
for idx in indexs:
    temp = [idx-4000, idx+4000]
    areas.append(temp)
```

```python
# Merge areas
merged_area = [areas[0]]
for area in tqdm(areas):
    # Check if there is intersection
    print("Merged:", merged_area)
    print("Area:", area)
    for i, item in enumerate(merged_area):
        if area[0]<=item[0]<=area[1]<=item[1]:
            temp = [area[0], item[1]]
            merged_area[i] = temp
        if item[0] <= area[0] <= item[1] <= area[1]:
            temp = [item[0], area[1]]
            merged_area[i] = temp
        if item[1] <= area[0]:
            merged_area.append(item)
        if area[1] <= item[0]:
            merged_area.append(item)
    print('=='*50)
```
