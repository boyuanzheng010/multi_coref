## Alignment

```python
# Load the season 1, eposide 1 data
with open('memor/data.json') as f:
    data = json.load(f)
keys = [item.strip().split('_') for item in data.keys()]

data_dict = defaultdict()
for item in keys:
    # Find the key of the season 1, eposide 1
    if item[0] == 'S01E01':
        idx = '_'.join(item)
        instance = data[idx]
        # deal with the instance
        starts = instance['seg_start']
        ends = instance['seg_end']
        sents = instance['sentences']
        speakers = instance['speakers']
        # Load every utterance
        for start, end, sent, speaker in zip(starts, ends, sents, speakers):
            print(float(start), type(start))
            print(float(end), type(end))
            print(sent)
            print(speaker)
            data_dict[(float(start), float(end))] = {
                'start': start,
                'end': end,
                'utterance': sent,
                'speaker': speaker
            }
            print('=='*50)
```

```python
# Load the caption
with open('captions/The.Big.Bang.Theory.S01E01.1080p.BluRay.x264-CiNEFiLE.ChsEngA.ass', 'r', encoding='utf-16') as f:
    doc = ass.parse(f)

caption_dict = defaultdict()

for x in doc.events:
    start = x.start.seconds + x.start.microseconds/1000000
    end = x.end.seconds + x.end.microseconds/1000000
    print(start)
    print(end)
    utterance = x.text.strip().split('\\N')
    # print(utterance)
    utt_zh = utterance[0].strip()
    print(utt_zh)
    utt_en = utterance[-1].strip().split('&}')[-1].strip()
    print(utt_en)
    print('=='*50)
    caption_dict[(start, end)] = {
        'start': start,
        'end': end,
        'utt_zh': utt_zh,
        'utt_en': utt_en
    }
```

```python
# Check Captions
caption_dict = caption_dict
caption_slots = tuple(caption_dict.keys())
caption_left = tuple([item[0] for item in caption_slots])
caption_right = tuple([item[1] for item in caption_slots])
temp = [float(0)]
for slot in caption_slots:
    temp.extend(slot)
caption_all = tuple(temp)

# Check Data
data_dict = data_dict
data_slots = tuple(data_dict.keys())
```

```python
def get_neighbor_times(time, time_points):
    for i in range(len(time_points)-1):
        if time_points[i] <= time <= time_points[i+1]:
            return (time_points[i], time_points[i+1])
```

```python
for data_slot in data_slots:
    utterance = data_dict[data_slot]
    utt_start = utterance['start']
    utt_end = utterance['end']
    # Locate start time
    start_times = get_neighbor_times(utt_start, caption_all)
    print(start_times)
    print(utterance)
    print(utt_start)
    print(utt_end)
    print('=='*50)
```
