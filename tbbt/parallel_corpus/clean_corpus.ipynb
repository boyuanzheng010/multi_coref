```python
def clean_sentence(text):
    text = text.strip().lstrip('-').lstrip().lstrip('.').lstrip()
    text_tokens = text.strip().split(" ")
    flag = False
    for token in text_tokens:
        if token.isupper():
            pass
            print(token)
    # print(text)
    # print(text_tokens)
```

```python
def clean_sentence_brackets(text):
    text = text.strip().lstrip('-').lstrip().lstrip('.').lstrip()
    new_text = re.sub(u"\\(.*?\\)|\\{.*?}|\\[.*?]", "", text)
    if len(text) != len(new_text):
        print(text)
        print(new_text)
        print("--"*50)
```

```python
with open('uncleaned_tbbt_en_zh.pkl', 'rb') as f_zh:
    zh_source = pkl.load(f_zh)
```

```python
for epi_key in zh_source:
    print(epi_key)
    scenes = zh_source[epi_key]
    for scene in scenes:
        for utt in scene:
            if 'en_subtitles' in utt:
                for item, zh in zip(utt['en_subtitles'], utt['zh_subtitles']):
                    text = item.strip().lstrip('-').lstrip().lstrip('.').lstrip()
                    text_token = text.split()
                    if "-" in text_token:
                        print(text_token, zh)
    print("=="*50)
```

```python
# Clean [], (), {}
output = {}
for epi_key in zh_source:
    scenes = zh_source[epi_key]
    for scene in scenes:
        scene_output = []
        for utt in scene:
            utt_output = deepcopy(utt)
            if 'en_subtitles' not in utt_output:
                continue
            for subtitle in utt_output['en_subtitles']:
                clean_sentence_brackets(subtitle)
            # print("=="*50)
```
