## Check Different Parsers

```python
# Load Stanza Parser
stanza_parser = spacy_stanza.load_pipeline('en')
```

```python
# Load Berkeley Parser
berkeley_parser = spacy.load('en_core_web_md')
berkeley_parser.add_pipe("benepar", config={"model": "benepar_en3"})

# berkeley_parser_np = spacy.load('en_core_web_md')
# berkeley_parser_np.add_pipe("benepar", config={"model": "benepar_en3"})
# berkeley_parser_np.add_pipe()
```

```python
# Spacy MD Parser
md_parser = spacy.load('en_core_web_md')
```

```python
# Load TRF Parser
trf_parser = spacy.load("en_core_web_trf")
```

```python
def get_spans(utt_, parser_):
    utt_ = utt_.replace("-", " ")
    noun_prons = []
    nps = []
    for i, item in enumerate(parser_(utt_)):
        pos = item.pos_
        # if pos in ["NOUN", "PRON"]:
        #     print(item.text, item.pos_)
        #     noun_prons.append((item.text, item.pos, i, i+1))
    for i, item in enumerate(parser_(utt_).noun_chunks):
        print(item)
        nps.append((item.text, item.start))
    # print(noun_prons)
    # print()
    # print(nps)
```

```python
for utt in utts:
    get_spans(utt, trf_parser)
    print("=="*50)
```

```python
# Use TRF Parser
for utt in utts:
    print(utt)
    for item in trf_parser(utt):
        print(item.text, item.pos_, item.dep_, item.has_dep())
    for item in trf_parser(utt.replace("-", " ")).noun_chunks:
        print(item)
    print("=="*50)
```

```python
# Use Stanza
for utt in utts:
    print(utt)
    utt = stanza_parser(utt)
    for token in utt:
        print(token.text, token.lemma_, token.pos_)
    print("=="*50)
```

```python
# Use Berkeley Parsing
for utt in utts:
    utt = berkeley_parser(utt.replace("-", " "))
    print(utt)
    # for item in utt:
    #     print(item, item.pos_)
    for item in utt.noun_chunks:
        print(item)
    print('=='*50)
```

```python
sentence = " So if a photon is directed through a plane with two slits in it and either slit is observed it will not go through both slits. If it’s unobserved it will, however, if it’s observed after it’s left the plane but before it hits its target, it will not have gone through both slits."
```

```python
result = {}

temp = []
for item in berkeley_parser(sentence):
    temp.append([item.text, item.pos_])
result['berkeley'] = temp

temp = []
for item in stanza_parser(sentence):
    temp.append([item.text, item.pos_])
result['stanza'] = temp

temp = []
for item in trf_parser(sentence):
    temp.append([item.text, item.pos_])
result['trf'] = temp

temp = []
for item in md_parser(sentence):
    temp.append([item.text, item.pos_])
result['md'] = temp
```

```python
for parser in result:
    print(parser, len(result[parser]))
    print(result[parser])
    print('=='*50)
```
