```python
import csv
import pickle as pkl
```

```python
transcript_list = []
with open('1_10_seasons_tbbt.csv', 'r') as f:
    csv_reader = csv.DictReader(f)
    for x in csv_reader:
        transcript_list.append(x)
```

```python
transcript_dict = {}
for x in transcript_list:
    episode_name = x['episode_name'].strip().split(' ')
    epi_sub = (int(episode_name[1]), int(episode_name[3]))
    if epi_sub not in transcript_dict:
        transcript_dict[epi_sub] = []
    dialogue = x['dialogue']
    person_scene = x['person_scene']
    transcript_dict[epi_sub].append([
        dialogue, person_scene
    ])
```

```python
# Check Transcript Scenes
for x in transcript_dict:
    count = 0
    for item in transcript_dict[x]:
        if item[-1]=='Scene':
            count += 1
    print(x, len(transcript_dict[x]), count)
```

```python
count = 0
for x in transcript_dict[(10,22)]:
    if x[-1]=='Scene':
        count += 1
        print("=====", x)
    else:
        print(x)
```
